2.   Explain overfitting and underfitting in ML. How can you detect and prevent them?
      -   overfitting means the model learns the training data too precisely—it memorizes patterns that are only true in the training set, including noise or random fluctuations. For detecting
      Training accuracy is high, but validation/test accuracy is noticeably lower.Validation loss starts rising while training loss keeps dropping.
      Learning curves for train and validation split apart.
      The steps of preventing are Cross-validation: Helps check if the model behaves consistently on different data splits.
      Regularization: e.g., L2, L1, dropout in deep learning. These penalize overly complex models.
      Early stopping: Stop training when validation performance stops improving.
      Simplify the model: Reduce depth, number of parameters, or features.
      Add more data or augment existing data.
      underfiting means The model is too simple to learn the underlying relationships in the data.It performs poorly on both training and test data.
      steps to detecting:
      Both training and validation accuracy are low.
      Training and validation curves stay close but never reach good performance.
      steps to prevent:Use a more expressive model more layers, more features, or more flexible algorithms.
      Reduce regularization if it is too strong.
      Train longer—sometimes the model just has not  learned enough.
      Improve feature engineering or add relevant features.
      Underfitting → High Bias, Low VarianceModel is too restricted and misses important patterns.
      Overfitting → Low Bias, High VarianceModel is too flexible and picks up random noise.

4. :What is an imbalanced dataset? Describe two techniques to handle it
(theoretical + practical).
    -   An imbalanced dataset is one where the number of samples in each class is not equal.
     One class (majority class) has many more examples, while the other class (minority class) has very few examples.
here is the therotical exolaination
SMOTE generates new synthetic examples of the minority class instead of simply duplicating existing ones.
It does this by:

        Picking a minority sample

        Finding its nearest neighbors

        Creating a new point somewhere between them

        This increases the minority class size and helps the model learn better patterns.



```python
import numpy as np
from sklearn.datasets import make_classification

# Generate a sample imbalanced dataset
X, y = make_classification(
    n_samples=1000,          # total number of samples
    n_features=10,           # total number of features
    n_informative=5,         # number of informative features
    n_redundant=0,           # number of redundant features
    n_repeated=0,            # number of repeated features
    n_classes=2,             # number of target classes
    n_clusters_per_class=1,  # number of clusters per class
    weights=[0.9, 0.1],      # make the dataset imbalanced (90% majority, 10% minority)
    flip_y=0,                # percentage of samples whose class is randomly flipped
    random_state=42
)

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
print(f"Class distribution in y: {np.bincount(y)}")
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(class_weight='balanced', solver='liblinear', random_state=42)
model.fit(X, y)
print("Model trained successfully with X and y.")
```

9. ) How do flight prices vary with the days left until departure? Identify any exponential price
surges and recommend the best booking window.
b)Compare prices across airlines for the same route (e.g., Delhi-Mumbai). Which airlines are
consistently cheaper/premium, and why?
Dataset: https://github.com/MasteriNeuron/datasets.git
(Include your Python code and output in the code box below.)
   -    

```python
import pandas as pd
import numpy as np

# Download the dataset from the GitHub repository
!wget https://raw.githubusercontent.com/MasteriNeuron/datasets/main/flight_prices.csv

# Load dataset
df = pd.read_csv("flight_prices.csv")

# Clean data
df = df.dropna(subset=["Price", "Days_Left", "Airline", "Source", "Destination"])

# -----------------------------------------
# PART A: Price variation vs Days Until Departure
# -----------------------------------------

avg_price_days = df.groupby("Days_Left")["Price"].mean().reset_index()

print("=== Average Price by Days Left ===")
print(avg_price_days.head(15))

# Identify exponential spikes:
# A simple method: % increase day-to-day
avg_price_days["pct_change"] = avg_price_days["Price"].pct_change() * 100

print("\n=== % Price Change (Day to Day) ===")
print(avg_price_days.tail(15))

# Find days where price jumps significantly (e.g., > 30%)
surge_days = avg_price_days[avg_price_days["pct_change"] > 30]
print("\nDays with Exponential Price Surge (>30% increase):")
print(surge_days)


# -----------------------------------------
# PART B: Compare airline prices for DELHI → MUMBAI
# -----------------------------------------

route_df = df[(df["Source"] == "Delhi") & (df["Destination"] == "Mumbai")]

airline_avg = route_df.groupby("Airline")["Price"].mean().sort_values()
print("\n=== Average Airline Prices for Delhi → Mumbai ===")
print(airline_avg)

print("\nCheapest Airline:", airline_avg.idxmin())
print("Most Expensive Airline:", airline_avg.idxmax())
```
